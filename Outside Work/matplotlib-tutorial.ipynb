```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


def calculate(name="2_1,2-DM-Adam",input_path=r"C:\Users\zanyi\OneDrive\Git hub\Python\Outside Work\DataBase.csv",
              defaulty=0,):
    df = pd.read_csv(input_path)
    names = df.iloc[:, [defaulty]].values.flatten().tolist()
    ret1 = df.iloc[:, [defaulty+5]].values.flatten().tolist()
    v1 = df.iloc[:, [defaulty+6]].values.flatten().tolist()
    ret2 = df.iloc[:, [defaulty+7]].values.flatten().tolist()
    v2 = df.iloc[:, [defaulty+8]].values.flatten().tolist()
    values=[(ret1[names.index(name)]+ret2[names.index(name)])/2]
    values.append(v1[names.index(name)])
    values.append(v2[names.index(name)])
    return values
def prepareBS10(compound="1_1,2-DM-Adam"):
    o=calculate(compound)
    ref=calculate(name="24_BS10")
    normalised = round(o[1]/ref[1]*ref[2])
    ratio = [o[0], round((normalised*100)/o[2])]
    return ratio


print(prepareBS10("1_1-M-Adam"))


```
